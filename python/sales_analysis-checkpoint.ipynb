```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Sales-Data-Analysis-Project.csv")

df.head()
```

```python
## understand the data
df.info()
```

```python
df.describe()
```

```python
##check missing value
df.isnull().sum()
```

```python
## clean
df["Postal Code"] = df["Postal Code"].fillna(0)
```

```python
df.isnull().sum()
```

```python
## total sales
total_sales = df["Sales"].sum()
print("Total Sales:", total_sales)
```

```python
## sales by region
region_sales = df.groupby("Region")["Sales"].sum()
region_sales
```

```python
## plotting graph
region_sales.plot(kind="bar")
plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.show()
```

```python
## save graph
plt.savefig("sales_by_region.png")
```

```python
## top 5 products
top_products = df.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head()
top_products
```

```python
## monthly sales trend
# convert to datetime 
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)

# extract month
df["Month"] = df["Order Date"].dt.to_period("M")

# group by month
monthly_sales = df.groupby("Month")["Sales"].sum()

# plot
monthly_sales.plot(figsize=(10,5))
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

```python
plt.savefig("monthly_sales_trend.png")
```

```python
## sales by category
category_sales = df.groupby("Category")["Sales"].sum()

category_sales.plot(kind="bar")
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
## sales by segment
segment_sales = df.groupby("Segment")["Sales"].sum()

segment_sales.plot(kind="bar")
plt.title("Sales by Segment")
plt.xlabel("Segment")
plt.ylabel("Total Sales")
plt.show()
```

```python
## top 10 customers
top_customers = df.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)

top_customers.plot(kind="barh")
plt.title("Top 10 Customers by Sales")
plt.xlabel("Total Sales")
plt.ylabel("Customer Name")
plt.gca().invert_yaxis()   
plt.show()
```

```python
## plt.gca().invert_yaxis() --- by usine this the highest one appears on top
## gca = get current axis (it tells matplotlib to give current graph axis so i can modify it)
## .invert_yaxis (the first appear on top, and last appear in last)
## in easy language do it in ascending order
```

```python
## sales by sub category
sub_category_sales = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False)

sub_category_sales.plot(kind="bar", figsize=(10,5))
plt.title("Sales by Sub-Category")
plt.xlabel("Sub-Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
## sales distribution (using seaborn)
import seaborn as sns
```

```python
sns.histplot(df["Sales"], bins=30, kde=True)

plt.title("Sales Distribution")
plt.xlabel("Sales Amount")
plt.ylabel("Frequency")
plt.show()
```

```python
## graph is heavily right skewed (means mostly sales are of low amount)
```

```python
## correlation heatmap
## step.1 creating shipping days column
df["Shipping Days"] = (pd.to_datetime(df["Ship Date"], dayfirst=True) -
    pd.to_datetime(df["Order Date"], dayfirst=True)
).dt.days
```

```python
## simplifying wht i did
##  Converted text dates into real date format
## then, Subtract Ship Date - Order Date
## then i Get number of days
```

```python
## applying correlation matrix
corr_matrix = df[["Sales", "Shipping Days"]].corr()
```

```python
## finally plotting heatmap
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
mean_sales = np.mean(df["Sales"])
median_sales = np.median(df["Sales"])
std_sales = np.std(df["Sales"])

print("Mean Sales:", mean_sales)
print("Median Sales:", median_sales)
print("Standard Deviation:", std_sales)
```
